```python
import numpy as np

#Precio, calificacion
producto_a = np.array ([20.0, 4.5])
print(f"Vector Producto: {producto_a}")
```

```python
print(f"Numero de ejes o dimensiones del producto: {producto_a.ndim}") #Un Array de una dimension, una simple secuencia o lista de numeros
print(f"forma del producto: {producto_a.shape}") #Es una estructura con un solo eje que tiene dos elementos
print(f"tamaño del producto: {producto_a.size}") #tamanho del vector o el numbero de componentes que contiene, o el numero de caracteristicas que contiene este vector.
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,6))
plt.quiver(0, 0, producto_a[0], producto_a[1], angles='xy', scale_units='xy', scale=1, color='red')
plt.title("producto a")
plt.xlabel("Precio")
plt.ylabel("Calificacion")
plt.xlim(0, 25)
plt.ylim(0, 10)
plt.grid(True, alpha=0.3)
plt.show()
```

```python
#4 Matrices: Ejenplo con calificaciones de Peliculas

Calificaciones = np.array([
    [5,1,2,0],
    [3,1,4,5],
    [2,4,2,0],
    [4,5,2,3],
])

print(Calificaciones)
```

```python
print(Calificaciones.ndim)
print(Calificaciones.shape)
print(Calificaciones.size)
```

```python
Calificaciones_usuario1 = Calificaciones[0]
print(Calificaciones_usuario1)

Calificaciones_pelicula3 = Calificaciones[:, 2]
print(Calificaciones_pelicula3)
```

```python
pixel_tensor = np.array([
    [
        [255, 0, 0],
        [0, 255, 0],
        [0, 0, 255]
    ], #primera "metamatriz"
[
        [255, 125, 100],
        [200, 255, 40],
        [115, 0, 255]
]#Segunda "Metamatriz"
])

print(pixel_tensor.ndim)
print(pixel_tensor.shape)
print(pixel_tensor.size)
```

```python
plt.imshow(pixel_tensor)
plt.title("Pixel Tensor")
plt.axis("off")
plt.show()
```
