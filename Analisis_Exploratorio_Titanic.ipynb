```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Cargamos los datos del Titanic (directo de la librería)
df = sns.load_dataset('titanic')

# 2. Vemos las primeras 5 filas para entender qué tenemos
print("--- PRIMER VISTAZO ---")
display(df.head())

# 3. Interrogatorio Estadístico Inicial
print("\n--- RESUMEN ESTADÍSTICO ---")
display(df.describe())
```

```python
#1. Cargar el dato y poner el nombre deseado
datos_del_barco = sns.load_dataset('titanic')

#2. Hay que darle ordenes usando tal nombre
#.head() significa "Muestrame la cabecera (las primeras filas)"
print(datos_del_barco.head())
```

```python
print("--%SUPERVIVENCIA POR GENERO ---")
print(datos_del_barco.groupby('sex')['survived'].mean())

print ("\n--- % SUPERVIVENCIA POR CLASE ---")
print(datos_del_barco.groupby('pclass')['survived'].mean())
```

```python
#1. Tamanho de los graficos
plt.figure(figsize=(10.,4))

#2. Grafico de Hombres vs Mujeres
plt.subplot(1,2,1) # (1 fila, 2 columnas, dibujo 1)
sns.barplot(data=datos_del_barco, x= 'sex', y = 'survived', palette= 'viridis')
plt.title("supervivencia por genero")

#3. grafico de Ricos vs Pobres.
plt.subplot(1,2,2) # (1 fila, 2 columnas, dibujo 2
sns.barplot(data=datos_del_barco, x= 'pclass', y = 'survived', palette= 'magma')
plt.title("supervivencia por Clase Social")

# Mostrar
plt.tight_layout()
plt.show()
```
